```python
#Reverse a number

num = int(input("enter number: "))
temp = num
rev = 0 
while num>0:
    rem = num%10
    rev=(rev * 10) + rem
    num = num //10
print("the given no is {} and reverse is {} ".format(temp,rev))
```
