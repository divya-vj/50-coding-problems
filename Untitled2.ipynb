```python
# String basics
s = "Hello World"
# Length
len(s) # 11
# Indexing — starts from 0
s[0] # "H"
s[-1] # "d" (last character)
s[-2] # "l" (second last)
# Slicing — [start:end:step] — end is NOT included
s[0:5] # "Hello"
s[6:] # "World"
s[:5] # "Hello"
s[::2] # "HloWrd" (every 2nd character)
s[::-1] # "dlroW olleH" (reversed)
# String methods — memorise ALL of these
s.upper() # "HELLO WORLD"
s.lower() # "hello world"
s.strip() # removes spaces from both ends
s.lstrip() # removes spaces from left only
s.rstrip() # removes spaces from right only
s.split() # ["Hello", "World"] — splits by space
s.split(",") # splits by comma
" ".join(["a","b"]) # "a b" — joins list into string
s.replace("l","x") # "Hexxo Worxd"
s.find("World") # 6 — index where found, -1 if not found
s.count("l") # 3 — how many times "l" appears
s.startswith("He") # True
s.endswith("ld") # True
s.isdigit() # False — checks if all chars are digits
s.isalpha() # False — checks if all chars are letters
s.isalnum() # False — checks if all chars are letters/digits
```
